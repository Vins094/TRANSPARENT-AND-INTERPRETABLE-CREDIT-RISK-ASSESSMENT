```python
import sys
import numpy as np
import pandas as pd
import sklearn
import xgboost
import seaborn as sns
import matplotlib
import lime
import shap
import torch
import skorch
import imblearn

# Function to get version or 'Not installed' if not available
def get_version(module):
    return module.__version__ if hasattr(module, '__version__') else 'Not installed'

# Print versions
print(f"Python version: {sys.version}")
print(f"NumPy version: {get_version(np)}")
print(f"imbalanced-learn version: {get_version(imblearn)}")
print(f"Pandas version: {get_version(pd)}")
print(f"scikit-learn version: {get_version(sklearn)}")
print(f"XGBoost version: {get_version(xgboost)}")
print(f"Seaborn version: {get_version(sns)}")
print(f"Matplotlib version: {get_version(matplotlib)}")
print(f"LIME version: {get_version(lime)}")
print(f"SHAP version: {get_version(shap)}")
print(f"PyTorch version: {get_version(torch)}")
print(f"Skorch version: {get_version(skorch)}")
```
