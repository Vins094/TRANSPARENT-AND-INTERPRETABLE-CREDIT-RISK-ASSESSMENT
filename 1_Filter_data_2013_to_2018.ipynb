```python
# Function to parse the date and check if it's in the desired range
#date_in_range
def parse_date(date_string):
    if pd.isna(date_string): #if the date in issue date column is na , return false
        return False
    try:

        if isinstance(date_string, float): # some dates comes with time in float form, such dates could be handled by converting
            date_string = f"{date_string:.2f}" #them to 2 decimal float and then removing float value and convert them to string

        date_string = date_string.split('.')[0] #and split them from decimal number to remove them
        date = datetime.strptime(date_string, '%b-%Y') #parse date in the format
        return 2013 <= date.year <= 2018 #return if the year of the date is from 2013 to 2018.
    except ValueError:
        print(f"Unable to parse date: {date_string}")
        return False  # return False otherwise
```

```python
# read loan csv in chunks
chunk_size = 100000  #
filtered_data = []
total_rows_processed = 0
total_rows_filtered = 0
```

```python
# File paths
loan_input_file_path = '/content/accepted_2007_to_2018q4.csv/accepted_2007_to_2018Q4.csv'
loan_output_file_path = '/content/accepted_2007_to_2018q4.csv/accepted_2013_to_2018_filtered.csv'

```

```python
#partially adapted from https://stackoverflow.com/questions/25962114/how-do-i-read-a-large-csv-file-with-pandas
try:
    # total rows in a csv file
    total_rows_csv = sum(1 for _ in open(loan_input_file_path, 'r', encoding='utf-8'))
    print(f"total_rows in csv: {total_rows_csv}")
    for chunk_n, c in enumerate(pd.read_csv(loan_input_file_path, chunksize=chunk_size, encoding='utf-8',
                             on_bad_lines='warn', low_memory=False, dtype=str)):

        if 'issue_d' in c.columns:
            c['issue_d'] = c['issue_d'].astype(str) #convert issue date in string
            filtered_chunk_value = c[c['issue_d'].apply(parse_date)]
            print(f"filtered_chunk in loan csv: {filtered_chunk_value.count()}")
            filtered_data.append(filtered_chunk_value)
            total_rows_filtered += len(filtered_chunk_value)
        else:
            print("Issue date column 'issue_d' not found in c , skipping the chunk")

        total_rows_processed += len(c)


        status_chunk = (total_rows_processed / total_rows_csv) * 100
        print(f"Status of chunk data processed: {status_chunk:.2f}%")
        print(f"Chunk {chunk_n + 1}: Processed {len(c)} rows, Filtered {len(filtered_chunk_value)} rows")
        print("Sample data from this chunk:")
        print(c['issue_d'].head())
        print("\n\n")

    # Concatenate the filtered chunks in a single dataframe
    if filtered_data:
        df = pd.concat(filtered_data, ignore_index=True)

        # save the data in new csv
        df.to_csv(loan_output_file_path, index=False, encoding='utf-8')

    else:
        print("No loan data processed")
        chunk_data_first = next(pd.read_csv(loan_input_file_path, chunksize=5, encoding='utf-8', dtype=str))
        print(chunk_data_first)

except Exception as e: # if any error occured
    print(f"Error occurred during processing: {str(e)}")
```
